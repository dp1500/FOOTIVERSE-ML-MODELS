# encoding formation column using 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df2[['gk rating', 'defence rating', 'midfield rating',
                                                'attack rating', 'net stamina', 'net physicallity','away gk rating', 'away defence rating', 'away midfield rating',
                                                'away attack rating', 'away net stamina', 'away net physicallity']] = scaler.fit_transform(df2[[ 'gk rating', 'defence rating', 'midfield rating',
                                                'attack rating', 'net stamina', 'net physicallity','away gk rating', 'away defence rating', 'away midfield rating',
                                                'away attack rating', 'away net stamina', 'away net physicallity']])
```

```python
df3 = df1
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()

df3[['gk rating', 'defence rating', 'midfield rating',
                                                'attack rating', 'net stamina', 'net physicallity','away gk rating', 'away defence rating', 'away midfield rating',
                                                'away attack rating', 'away net stamina', 'away net physicallity']] = mms.fit_transform(df3[[ 'gk rating', 'defence rating', 'midfield rating',
                                                'attack rating', 'net stamina', 'net physicallity','away gk rating', 'away defence rating', 'away midfield rating',
                                                'away attack rating', 'away net stamina', 'away net physicallity']])
```

```python
df3.head()
```

```python
y = df3[['home goals','away goals']].values
x = df3.drop(['home goals', 'away goals', 'home_result_list'], axis=1).values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.05, random_state=42)
```

#MODEL BUILDING USING LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
linearReg = LinearRegression()
linearReg.fit(X_train,y_train)
```

```python
predictions = linearReg.predict(X_test)
predictions
```

```python
print(predictions)
```

```python
# a = np.array([[1.5, 1.4],
#               [2.1, 2.5]])
# round_off_values = np.around(a)
# print(round_off_values)
```

##checking accuracy without round off

```python
predictions_result = []
for i in range(len(predictions)):
  if predictions[i][0] - predictions[i][1] > 0.3:
    predictions_result.append('W')
  elif predictions[i][0] - predictions[i][1] < -0.3:
    predictions_result.append('L')
  elif predictions[i][0] - predictions[i][1] <= 0.3 and predictions[i][0] - predictions[i][1] >= -0.3:
    predictions_result.append('D')

predictions_result = np.array(predictions_result)
```

```python
predictions_result
```

```python
test_results = []
for i in range(len(predictions)):
  if y_test[i][0] - y_test[i][1] > 0:
    test_results.append('W')
  elif y_test[i][0] - y_test[i][1] < 0:
    test_results.append('L')
  else:
    test_results.append('D')

test_results = np.array(test_results)
```

```python
test_results
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

le.fit(test_results)

predictions_result = le.transform(predictions_result)
test_results = le.transform(test_results)
```

## metrics

```python
from sklearn.metrics import accuracy_score
accuracy_score(test_results, predictions_result)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

le.fit(test_results)

rounded_off_predictions_result = le.transform(rounded_off_predictions_result)
test_results = le.transform(test_results)
```
